```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d={'C':[12,20,34,45,55],'C++':[15,25,36,50,60],'Java':[18,30,40,55,70],'R':[20,35,45,60,80],'Python':[25,40,53,75,100]}
file_sizes=[200,400,600,800,1000]
times_c = d["C"]
print(times_c)
times_cpp = d["C++"]
times_java =d["Java"]
times_r = d["R"]
times_python = d["Python"]
df=pd.DataFrame(d,index=['200MB','400MB','600MB','800MB','1000MB'])
plt.plot(file_sizes, times_c, marker='o', label='C')
plt.plot(file_sizes, times_cpp, marker='s', label='C++')
plt.plot(file_sizes, times_java, marker='^', label='Java')
plt.plot(file_sizes, times_r, marker='d', label='R')
plt.plot(file_sizes, times_python, marker='x', label='Python')
print(df)
plt.title('Performance Comparison: Converting Text Files to Upper Case')
plt.xlabel('File_Size')
plt.ylabel('Time Taken (seconds)')
plt.legend()
plt.show()
```
